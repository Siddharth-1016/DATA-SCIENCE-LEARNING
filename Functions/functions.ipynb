```python
#Why functions?
#-code reusability -code readability


def even_or_odd(num):
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")

even_or_odd(5)
even_or_odd(6)
```

```python
#function with multple parameter
def add(a,b):
    return a+b

def sub(a,b):
    return a-b

print(add(6,7))
print(sub(7,6))
```

```python
##Default parameter

def greet(name = "Siddharth"):
    print(f"Hello {name} welcome !")

greet("Krish")
greet()
    
```

```python
##Variable length Arguments
##Positional and Keywords arguments

def print_number(*args):
    for number in args:
        print(number, end=" ")

print_number(1,2,3,4,5,6)
```

```python
#Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")

print_details(name="Krish",age=21,gender="Male")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument: {val}")

    for key, value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,name="Sid",age=25)
```

```python
def calculate_total_cost(**kwargs):
    ans = 0
    for key, val in kwargs.items():
        print(f"{key}:{val}")
        ans += val

    return ans

result = calculate_total_cost(Coffee=250,Tea=200)
print(f"Total cost = {result}")
```
