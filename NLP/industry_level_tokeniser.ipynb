```python
import re
import nltk
import emoji
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

## FUNCTION TO EXTRACT EMOJIS

```python
def extract_emojis(text):
    return [c for c in text if c in emoji.EMOJI_DATA]
```

# FUNCTION TO CLEAN TEXT

```python
def clean_text(text):
    
    # lowercase
    text = text.lower()
    
    # remove urls
    text = re.sub(r'http\S+|www\S+', ' ', text)
    
    # remove emails
    text = re.sub(r'\S+@\S+', ' ', text)
    
    # remove currency and numbers
    text = re.sub(r'₹?\d+(,\d+)*', ' ', text)
    
    # split hashtags (#bestphoneever → bestphoneever)
    text = re.sub(r'#', '', text)
    
    return text
```

# PIPELINE FOR TOKENIZATION

```python
def industry_tokenizer(text):

    #extract emojis first
    emojis = extract_emojis(text)

    #clean text
    text = clean_text(text)

    #tokenize words
    words = word_tokenize(text)

    #keep only alphabets
    words = [w for w in words if w.isalpha()]

    #combine words + emojis
    final_tokens = words + emojis

    return final_tokens
```

```python
review = "Omg!!! This phone is sooo goood 😍🔥🔥!!! Worth ₹30,000?? Visit www.shop.com #BestPhoneEver"

tokens = industry_tokenizer(review)
print(tokens)
```

```python
reviews = [
    "Loved it 😍 Best purchase ever!!!",
    "Battery sucks 😡 Waste of money",
    "Worth ₹25,000 🔥🔥 #AwesomePhone"
]

tokenized_reviews = [industry_tokenizer(r) for r in reviews]

for t in tokenized_reviews:
    print(t)
```
