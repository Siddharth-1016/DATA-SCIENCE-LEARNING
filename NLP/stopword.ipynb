```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('french')
```

```python
paragraph = '''**A. P. J. Abdul Kalam** was one of India’s most beloved scientists and leaders, widely known as the “Missile Man of India.” Born in 1931 in Rameswaram, Tamil Nadu, he came from a humble background and sold newspapers as a child to support his family. Despite financial challenges, his passion for learning and dedication to education helped him pursue aerospace engineering and build a remarkable career in science and technology. Kalam played a key role in India’s missile and nuclear programs, contributing significantly to the country’s defense and space research. His work with organizations like ISRO and DRDO made him a national hero and an inspiration for millions of young Indians.
In 2002, he became the 11th President of India and earned the title “People’s President” because of his simplicity, kindness, and strong connection with students and youth. He believed that education, innovation, and hard work were the pillars of national development. Kalam often encouraged young people to dream big and work tirelessly to achieve their goals, famously saying, “Dream, dream, dream. Dreams transform into thoughts and thoughts result in action.” Even after his presidency, he continued teaching and inspiring students until his last breath in 2015. His life remains a powerful example of humility, perseverance, and dedication to the nation.
'''
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer = SnowballStemmer('english')
```

```python
#apply stopwords and filter and then apply stemming

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```
