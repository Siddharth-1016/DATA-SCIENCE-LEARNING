```python
words = ["eating","eaten","eats","writting","writes","programming","programs","histroy","finally","finalized"]
```

# PORTER STEMMER

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + " ----> " + stemming.stem(word))
```
