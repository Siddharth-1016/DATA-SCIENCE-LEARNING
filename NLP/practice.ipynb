```python
corpus = '''Natural Language Processing is one of the most exciting fields in Artificial Intelligence. Every day, millions of people interact with chatbots, search engines, and voice assistants without realizing how much text processing happens behind the scenes. From analyzing customer reviews to detecting spam emails, NLP helps computers understand human language and make smarter decisions.'''
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(corpus)

for sentence in sentences:
    print(sentence)
```

```python
from nltk.tokenize import word_tokenize

for sentence in sentences:
    words = word_tokenize(sentence)
    print(words)
```

```python
from nltk.stem import SnowballStemmer

snowstemmer = SnowballStemmer("english")
```

```python
stemmed_words = []

for word in words:
    stemmed_words.append(snowstemmer.stem(word))

print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
for word in words:
    print(word+"------>"+lemmatizer.lemmatize(word))
```
