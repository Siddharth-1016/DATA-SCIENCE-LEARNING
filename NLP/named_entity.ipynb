```python
import nltk

# 1. Download all required NLTK data packages (including the missing one)
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')  # <--- The fix for the LookupError

# Optional: Append custom NLTK data path if your environment requires it
# nltk.data.path.append("C:/Users/HP/AppData/Roaming/nltk_data")

# 2. Define the text
sentence = '''The Eiffel Tower was built in 1887–1889 as the centerpiece of the 1889 World’s Fair celebrating the French Revolution’s centenary in Paris. Designed by engineer Gustave Eiffel, it was initially criticized by artists and writers who thought it ruined the city’s beauty and was meant to stand only for 20 years. However, its usefulness for radio communication and military purposes during World War I saved it from demolition. Over time, the tower became a global icon of France, attracting millions of visitors every year and symbolizing innovation, romance, and architectural brilliance.'''

# 3. Tokenize the sentence into words
words = nltk.word_tokenize(sentence)

# 4. Perform Part-of-Speech (POS) tagging
tag_elements = nltk.pos_tag(words)

# 5. Perform Named Entity Recognition (NER) chunking
ner_tree = nltk.ne_chunk(tag_elements)

# 6. Output the results
# Print the tree structure directly to the console/notebook
print(ner_tree)

# If you are running locally and want to open the graphical tree window, uncomment the line below:
# ner_tree.draw()
```
