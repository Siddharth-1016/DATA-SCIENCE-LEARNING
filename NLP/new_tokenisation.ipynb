##TOKENISING PARAGRAPH INTO SENTENCES

```python
from nltk.tokenize import sent_tokenize

text = "I love AI. NLP is amazing. Python is easy. This is my first NLP notebook."

documents = sent_tokenize(text)

print(documents)
```

```python
for sentence in documents:
    print(sentence)
```

TOKENISING SENTENCES INTO WRODS
TOKENISING PARAGRAPHS INTO WORDS

```python
from nltk.tokenize import word_tokenize
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(text)
```
