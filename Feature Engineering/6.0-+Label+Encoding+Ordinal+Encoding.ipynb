LABEL ENCODER

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({'Diagnosis':['Healthy','Sick','Healthy','Critical','Sick']})
```

```python
lb = LabelEncoder()
```

```python
df['Diagnonis Encoded'] = lb.fit_transform(df['Diagnosis'])
```

```python
print(df)
```

ORDINAL ENCODING

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({'Performance':['Average','Poor','Good','Excellent','Average']})
```

```python
performance_order = ['Poor','Average','Good','Excellent']
```

```python
oe = OrdinalEncoder(categories=[performance_order])
```

```python
df['Performance Encoded'] = oe.fit_transform(df[['Performance']])
```

```python
print(df)
```
