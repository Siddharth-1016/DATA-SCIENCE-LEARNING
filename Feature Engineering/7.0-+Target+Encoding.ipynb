```python
import pandas as pd

# Sample data: 'City' is the feature, 'House_Price' is the target
data = pd.DataFrame({
    'City': ['Delhi', 'Mumbai', 'Pune', 'Delhi', 'Mumbai', 'Pune', 'Delhi'],
    'House_Price': [50, 100, 40, 60, 120, 45, 55] # Prices in some unit
})

# 1. Calculate the mean house price for each city
city_means = data.groupby('City')['House_Price'].mean()

# 2. Sort the cities by their mean price
sorted_cities = city_means.sort_values().index

# 3. Create a dictionary to map the sorted cities to ordinal numbers
# Pune (lowest mean) -> 0, Delhi -> 1, Mumbai (highest mean) -> 2
ordinal_mapping = {city: i for i, city in enumerate(sorted_cities)}

# 4. Map the values back to the dataframe
data['City_Encoded'] = data['City'].map(ordinal_mapping)

print("Mean Prices:\n", city_means, "\n")
print("Mapping Dictionary:\n", ordinal_mapping, "\n")
print(data)
```
