```python
##DATA MANIPULATION WITH DATAFRAME
data = {
    'Name':['Krish','Siddharth','Anshul','Yash'],
    'Age':[30,22,21,23],
    'City':['Mumbai','Jabalpur','Jhansi','Indore']
}

df = pd.DataFrame(data)
```

```python
##ADDING A COLUMN
df['Salary']=[100000,200000,150000,100000]

df
```

```python
##REMOVING A COLUMN

df.drop('Salary',axis=1,inplace=True)
df
```

```python
##ADDING THE AGE TO THE COLUMN
df['Age'] = df['Age']+1
print(df)
print("\n")

```

```python
##DROPING A PARTICULAR ROW

age_even = df[df['Age']%2 == 0].index

df.drop(age_even,inplace=True)
df
```
