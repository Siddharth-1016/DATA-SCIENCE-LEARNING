```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# 1. Load the built-in tips dataset from seaborn
df = sns.load_dataset('tips')

# 2. Separate Features (X) and Target (y)
# We want to predict 'time', so we drop it from our features
X = df.drop('time', axis=1)
y = df['time']

# 3. Preprocess Categorical Variables
# Scikit-learn models need numbers. We'll use pd.get_dummies for quick one-hot encoding.
# drop_first=True prevents the "dummy variable trap" (perfect collinearity)
X_encoded = pd.get_dummies(X, drop_first=True)

# 4. Split the data into Training and Testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# 5. Initialize and Train the Naive Bayes Classifier
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# 6. Make Predictions on the test set
y_pred = nb_model.predict(X_test)

# 7. Evaluate the results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
