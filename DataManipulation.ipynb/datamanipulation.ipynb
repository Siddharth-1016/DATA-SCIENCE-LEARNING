```python
#Handling missing values
df.isnull()

#where there is false - it means no missing values
#if true - it means missing values
```

```python
#count of missing values in each column
df.isnull().sum()
```

```python
##Filling all the missing values with zeros
# df_filled = df.fillna(0)

df['Unit Price'].fillna(df['Unit Price'].mean())
df
```

```python
##Remaining Columns
df.rename(columns={'Date':'Sales Date'},inplace=True)
df.head()
```

```python
#Change Datatype
df['Unit Price'].astype(int)
df
```

```python
df['New Unit Price'] = df['Unit Price'].apply(lambda x:x*2)
df
```

```python
##Data Aggregating and Grouping
df.head(5)
```

```python
total_product_category_sum = df.groupby(['Product Category','Region'])['Units Sold'].sum()
print(total_product_category_sum)

print("\n")

grouped_mean = df.groupby(['Product Category','Region'])['Unit Price'].mean()
print(grouped_mean)
```

```python
grouped_Agg = df.groupby('Product Category')['Unit Price'].agg([
    ('Mean Price', 'mean'),
    ('Total Cost', 'sum'),
    ('Count of Products', 'count')
])

print(grouped_Agg)
```
