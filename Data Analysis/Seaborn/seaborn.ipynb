```python
import seaborn as sns
```

```python
##Basic Plotting With Seaborn
tips = sns.load_dataset('tips')
tips
```

1. Scatterplot
Significance: A scatterplot displays the relationship between two continuous variables. It helps visualize the correlation (positive, negative, or none) and the strength of that relationship.

Use Cases:

Identifying potential correlations (e.g., as years of experience increase, salary tends to increase).

Detecting outliers that deviate from the general pattern.

Evaluating if a linear or non-linear model would be appropriate for prediction.

```python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

2. Line Plot
Significance: A line plot connects data points, usually to show how a continuous variable changes in relation to another continuous variable, most commonly time.

Use Cases:

Visualizing trends over time (e.g., stock prices, temperature changes, sales performance month-over-month).

Comparing the performance of multiple series on the same metric.

Highlighting cyclical patterns or seasonal variations.

```python
##Line Plot

sns.lineplot(x='size', y='total_bill', data = tips)
plt.title("Line Plot of Size vs Total Bills")
plt.show()
```

3. Bar Plot (or Bar Chart)
Significance: A bar plot represents categorical data with rectangular bars. The height (or length) of the bars is proportional to the values they represent, which can be counts, sums, averages, etc.

Use Cases:

Comparing the counts or frequencies of different categories (e.g., number of students in each major).

Comparing a single metric across several discrete groups (e.g., average spending by different customer segments).

Grouped or Stacked Bar Charts can show the distribution of a subcategory within the main categories.

```python
##Cateogiral Plots
##Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Plot of Day vs Total Bills")
plt.show()
```

```python
##Pie Chart
grouped_data = tips.groupby('sex')['total_bill'].sum()
labels = grouped_data.index
sizes = grouped_data.values
colors = ['lightblue','lightgreen']
explode = (0,0)
plt.figure(figsize=(6,6))
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",startangle=140)
plt.title('Total Bill by Sex')
plt.show()
```

```python
#Box Plot
sns.boxplot(x='day',y='total_bill',data = tips)
plt.title("Box plot of Total Bill per day")
plt.show()
```

```python
#Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
##Histograms
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

```python
sns.kdeplot(tips['total_bill'],shade = True)
```

```python
##Pair Plot - relation between each variable
sns.pairplot(tips)
```
