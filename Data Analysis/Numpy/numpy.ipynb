```python
arr4 = np.array([[1,2,3,4],[5,6,7,8]])

print(arr4)
print(arr4.shape)
print(arr4.ndim)
print(arr4.size)
print(arr4.dtype)
print(arr4.itemsize)
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([10,20,30,40,50])

print("Addition:",arr1+arr2)
print("Substraction:",arr2-arr1)
print("Multiplication:",arr1*arr2)
print("Divisional:",arr2/arr1)
```

```python
##Universal function
arr = np.array([2,3,4,5,6])

##square root
print(f"Square root: {np.sqrt(arr)}")

##Exponential
print(f"Exponential: {np.exp(arr)}")

##Sine
print(f"Sine function: {np.sin(arr)}")

##Natural log
print(f"Natural log: {np.log(arr)}")
```

```python
data = np.array([1,2,3,4,5])

##calculate the mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

normalized_data = (data-mean)/std_dev
print("Normalised data: ", normalized_data)
```

```python
##Logical Operation

data = np.array([1,2,3,4,5,6,7,8])

print(data[data>5])
print(data[(data>=5) & (data<=8)])
```
