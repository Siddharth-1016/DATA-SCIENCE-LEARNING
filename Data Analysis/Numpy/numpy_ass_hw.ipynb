### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.
2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

```python
arr_51 = np.random.randint(1,25,size=(3,4))
arr_52 = np.random.randint(1,25,size=(3,4))

print("\n Addition \n")
print(arr_51+arr_52)
print("\n Subtraction \n")
print(arr_51-arr_52)
print("\n Multiplication \n")
print(arr_51*arr_52)
print("\n Division \n")
print(arr_51/arr_52)



```

```python
arr6 = np.random.randint(1,25,size=(4,4))
print(arr6)
rows, columns = arr6.shape
print("\nSum of Rows\n")
for i in range(rows):
    sum = 0
    for j in range(columns):
        sum += arr6[i,j]
    print(f"{i}th row sum = {sum}")

print("\nSum of Columns\n")
for i in range(columns):
    sum = 0
    for j in range(rows):
        sum += arr6[j,i]
    print(f"{i}th col sum = {sum}")

# Compute the row-wise and column-wise sum
row_sum = np.sum(arr6, axis=1)
column_sum = np.sum(arr6, axis=0)

print("\nRow-wise sum:")
print(row_sum)
print("\nColumn-wise sum:")
print(column_sum)

    

```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).


```python
arr7 = np.random.randint(1,100,size=(5,5))

print(f"Mean of array = {np.mean(arr7)}\n")
print(f"Standard deviation = {np.std(arr7)}\n")
print(f"Variance = {np.var(arr7)}\n")

normalised_Array = (arr7-np.mean(arr7))/(np.std(arr7))
print(f"Normalised Array : \n{normalised_Array}")
```

### Assignment 5: Broadcasting

1. Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.
2. Create a NumPy array of shape (4, 4) filled with random integers. Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.

```python
arr8 = np.random.randint(1,100,size=(3,3))
arr9 = np.random.randint(1,9,size=(3,))
print(f"Original 2d array: \n{arr8}\n")
print(f"1d array : \n{arr9}\n")

resulted_array = arr8 + arr9

print(f"Manipulated added 2d array: \n{resulted_array}\n")

subtracted_array = arr8 - arr9

print(f"Manipulated subtracted 2d array: \n{subtracted_array}")

```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

```python
import numpy as np
from numpy import linalg as la
arr10 = np.random.randint(1,100,size=(3,3))
print(arr10)

determinant = la.det(arr10)
print(f"\nDeterminant = {determinant}\n")

inverse = la.inv(arr10)
print(f"\nInverse = {inverse}\n")

eigen_val, eigen_vector = la.eig(arr10)
print(f"\nEigen Value = {eigen_val.round(4)}\n")
```
